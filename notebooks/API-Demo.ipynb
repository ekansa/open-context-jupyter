# First we will import a Python library that makes Web requests

```python
import requests
```

Now let's make a request to a URL

```python
url = 'https://opencontext.org/query/?obj=getty-aat-300265183'
headers = {
    'accept': 'application/json',
    'User-Agent': 'oc-api-client',
}
r = requests.get(url=url, headers=headers)
r.json()
```
